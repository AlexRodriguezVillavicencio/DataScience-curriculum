## $$\text{Sistemas de recomendación basado en el algoritmo de Apriori}$$

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('../data/Market_Basket_Optimisation.csv', header=None)
dataset
```

Antecendetes: Una tienda al sur de Francia, se busca optimizar la ubicación de los diferentes productos para generar más ventas. Por tanto se busca conocer las reglas de asociación de los diferentes productos. 
La tienda es un lugar popular y concurrido, además de ser de reunión.

Cada observación (fila) es un cliente que lleva los diferentes productos.

### Preprocesamos los datos para poder trabajarlos con el algoritmo

```python
transactions = []

for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

### Entrenamos el algoritmo apriori

```python
from apyori import apriori
rules = apriori(transactions, min_support = 0.003 , min_confidence = 0.2,
                min_lift = 3, min_length = 2)
```

### visualizamos

```python
results = list(rules)

print(results[4])
```
