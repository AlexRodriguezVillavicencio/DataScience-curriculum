## $$\text{Clustering Jérarquico}$$

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

### Importamos el data set

```python
dataset = pd.read_csv('../data/Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

### Técnica del dendrograma

```python
import scipy.cluster.hierarchy as sch 

dendogram = sch.dendrogram(sch.linkage(X, method="ward"))
plt.title("Dendrograma")
plt.xlabel('clientes')
plt.ylabel('distancia euclidea')
plt.show()
```

### Ajustar el clustering jerárquico

```python
from sklearn.cluster import AgglomerativeClustering 

hc = AgglomerativeClustering(n_clusters=5, affinity="euclidean",linkage="ward")
y_hc = hc.fit_predict(X)
y_hc
```

### Representación gráfica

```python
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1], s=30, c="red", label="cautos")
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1], s=30, c="purple", label="estandar")
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,1], s=30, c="green", label="objetivo")
plt.scatter(X[y_hc == 3,0], X[y_hc == 3,1], s=30, c="cyan", label="descuidados")
plt.scatter(X[y_hc == 4,0], X[y_hc == 4,1], s=30, c="orange", label="conservadores")

plt.xlabel('ingresos anuales (en miles de dólares)')
plt.ylabel('puntuación de gastos (1-100)')
plt.legend()
plt.show()
```
