## $$\text{Arboles de Regresión}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../data/Position_Salaries.csv')
x = df.iloc[:,1].values.reshape(-1,1)
y = df.iloc[:,2].values.reshape(-1,1)
```

### Ajustamos el modelo

```python
from sklearn.tree import DecisionTreeRegressor 

reg = DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

### Predecimos

```python
y_pred = reg.predict([[6.5]])
y_pred
```

```python
y_pred = reg.predict([[6.5]])
y_pred
```

### Visualizamos

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color="red")
plt.show()
```

### Veamos el lado oculto si llenamos con más datos

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(-1,1)
plt.scatter(x,y)
plt.plot(x_grid,reg.predict(x_grid),color="red")
plt.show()
```

Podemos ver que el algoritmo realiza cortes medios entre cada año, así se da un mismo sueldo entre 5.5 - 6.5 por ejemplo.
