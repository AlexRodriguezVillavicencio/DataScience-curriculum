## $$\text{Bosques Aleatorios}$$
$$\text{Inspirado de la sabiduria de masas}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/Position_Salaries.csv')
x = df.iloc[:,1].values.reshape(-1,1)
y = df.iloc[:,2].values.reshape(-1,1)
```

### Ajustamos el modelo

```python
from sklearn.ensemble import RandomForestRegressor 

reg = RandomForestRegressor(n_estimators=10, random_state=0)
reg = reg.fit(x,y.ravel())
```

### Predecimos

```python
# Un talento menciona que gana 160 000 por los seis a√±os y medio
y_pred = reg.predict([[6.5]])
y_pred
```

### Visualizamos

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape(-1,1)
plt.scatter(x,y)
plt.plot(x_grid,reg.predict(x_grid),color="red")
plt.show()
```

Podemos ver que los √°rboles y bosques son discretos
